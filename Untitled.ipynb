```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
```

```python
def crosses_line(y_end1, y_end2, t):
    lines = [i*t for i in range(5)]
    for line in lines:
        if line > y_end1 and line < y_end2:
            return True
        if line < y_end1 and line > y_end2:
            return True
        
    return False
```

```python
def needles(t, l): 
    # random point on the board
    x = rand.uniform(-4, 4)
    y = rand.uniform(0, t*4)

    # end points of the line
    y_end1 = y + (l/2)
    
    y_end2 = y - (l/2)
    
    return x, y, y_end1, y_end2, crosses_line(y_end1, y_end2, t)
```

```python
# l = length of needle
# t = distance between parallel lines
# N = number of needles
# plot = when you want to see the plot or not. Default set to True.
def buffon(l, t, N, plot=True):
    # create a board    
    fig = plt.figure(figsize=(9,7))
    
    board = fig.add_subplot()
    
    for i in range(5):
        plt.axhline(y=i*t, color="black")
    
    cross = 0
    
    # place the needles on the board
    for i in range(N):
        x, y, y_end1, y_end2, crossed_line = needles(t, l)
        if crossed_line:
#             board.plot([x,y_end1], [x,y_end2], 'g')
            c = plt.Circle((x,y), l/2, fill=None, color='g')
            board.add_patch(c)
            cross += 1
        else:
#             board.plot([x,y_end1], [x,y_end2], 'b')
            c = plt.Circle((x,y), l/2, fill=None, color='b')
            board.add_patch(c)
    
    board.axis('equal')
    
    if not plot:     
        plt.close()
                        
    # calculate the observed, esimated probability, Pi aproximate
    observed_prob = cross/N
    
    if l > t:
        estimated_prob = 1+(2/np.pi)*((l/t) * (1 - np.sqrt(1 - (t**2 / l**2))) - np.arcsin(t/l))
        aprox_pi = (2 * N/cross)*((l - np.sqrt(l**2 - t**2))/t + np.arccos(t/l))
    else:
        estimated_prob = (2*l)/(t*np.pi)
        aprox_pi = (2*l*N)/(t*cross)
    
    
    text = ('Observed Probability = %s , Estimated Probability = %f , Aproximated Pi = %f' 
            % (observed_prob, estimated_prob, aprox_pi))
    fig.text(.12,.9,text)
    
    return aprox_pi
```

```python
buffon(1,2,10000)
```
